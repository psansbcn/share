```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load Barcelona weather_data
df=pd.read_csv('barcelona_rain.csv',index_col=[0])
df=df[['fecha','sol','tmed','tmed-1','tmed-2','tmed-3','presMax','presMax-1','presMax-2','presMax-3','velmedia','velmedia-1','velmedia-2','velmedia-3','dir','dir-1','dir-2','dir-3','humidity','humidity-1','humidity-2','humidity-3','prec-1','prec-2','prec-3','prec']]
df.head()
```

```python
dfmain=df[['fecha','tmed','sol','presMax','velmedia','dir','humidity','prec']]
dfmain.head(3)
```

```python
fig, ax = plt.subplots(figsize=(19,2))
sns.boxplot(x=dfmain["prec"], whis=[5, 95])
plt.savefig('boxplot.jpg', scale=4)
```

```python

#ax[0].set(xlim=(0, 500000))

fig, ax = plt.subplots(1,3,figsize=(19,4))
sns.histplot(x='tmed', data=dfmain, color='g', kde=True , edgecolor="w", alpha=.3, ax=ax[0],
             bins=300)
ax[0].set()

sns.histplot(x='sol', data=dfmain, color='y', kde=True , edgecolor="w", alpha=.3, ax=ax[1],
             bins=300)
ax[1].set(ylim=(0, 300))

sns.histplot(x='presMax', data=dfmain, color='r', kde=True , edgecolor="w", alpha=.3, ax=ax[2],
             bins=300)
ax[2].set()
```

```python
fig, ax = plt.subplots(1,3,figsize=(19,4))
sns.histplot(x='velmedia', data=dfmain, color='pink', kde=True , edgecolor="w", alpha=.3, ax=ax[0],
             bins=100)
ax[0].set()

sns.histplot(x='dir', data=dfmain, color='violet', kde=True , edgecolor="w", alpha=.3, ax=ax[1],
             bins=100)
ax[1].set()

sns.histplot(x='humidity', data=dfmain, color='cyan', kde=True , edgecolor="w", alpha=.3, ax=ax[2],
             bins=100)
ax[2].set()
```

```python
fig, ax = plt.subplots(figsize=(19,4))
sns.histplot(x='prec', data=dfmain, color='b', kde=True , edgecolor="w", alpha=.3, ax=ax,
             bins=50)
```

```python
fig, ax = plt.subplots(figsize=(9,4))
sns.violinplot(x=dfmain["prec"], color='b')

plt.savefig('violin.jpg', scale=4)
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax = sns.regplot(x=dfmain['presMax'], y=dfmain['prec'], color="violet", marker=".", scatter_kws={'alpha':0.4},
                y_jitter = 1, x_jitter = 1)
```

```python
fig, ax = plt.subplots(1,3,figsize=(19,4))

sns.regplot(x=dfmain['dir'], y=dfmain['prec'], color="violet", marker=".", scatter_kws={'alpha':0.2}, ax=ax[0])
ax[0].set()

sns.regplot(x=dfmain['velmedia'], y=dfmain['prec'], color="pink", marker=".", scatter_kws={'alpha':0.2}, ax=ax[1])
ax[1].set()

sns.regplot(x=dfmain['presMax'], y=dfmain['prec'], color="r", marker=".", scatter_kws={'alpha':0.1}, ax=ax[2])
ax[2].set()
```

```python
fig, ax = plt.subplots(1,3,figsize=(19,4))

sns.regplot(x=df['dir-1'], y=df['prec'], color="violet", marker=".", scatter_kws={'alpha':0.2}, ax=ax[0])
ax[0].set(ylim=(-3, 15))

sns.regplot(x=df['dir-2'], y=df['prec'], color="pink", marker=".", scatter_kws={'alpha':0.2}, ax=ax[1])
ax[1].set(ylim=(-3, 15))

sns.regplot(x=df['dir-3'], y=df['prec'], color="r", marker=".", scatter_kws={'alpha':0.2}, ax=ax[2])
ax[2].set(ylim=(-3, 15))

```

```python
fig, ax = plt.subplots(1,3,figsize=(19,4))

sns.regplot(x=df['presMax-1'], y=df['prec'], color="r", marker=".", scatter_kws={'alpha':0.1}, ax=ax[0])
ax[0].set(ylim=(-3, 15))

sns.regplot(x=df['presMax-2'], y=df['prec'], color="r", marker=".", scatter_kws={'alpha':0.1}, ax=ax[1])
ax[1].set(ylim=(-3, 15))

sns.regplot(x=df['presMax-3'], y=df['prec'], color="r", marker=".", scatter_kws={'alpha':0.2}, ax=ax[2])
ax[2].set(ylim=(-3, 15))

```

```python
fig, ax = plt.subplots(1,3,figsize=(19,4))

sns.regplot(x=df['humidity-1'], y=df['prec'], color="cyan", marker=".", scatter_kws={'alpha':0.1}, ax=ax[0])
ax[0].set(ylim=(-3, 15))

sns.regplot(x=df['humidity-2'], y=df['prec'], color="cyan", marker=".", scatter_kws={'alpha':0.1}, ax=ax[1])
ax[1].set(ylim=(-3, 15))

sns.regplot(x=df['humidity-3'], y=df['prec'], color="cyan", marker=".", scatter_kws={'alpha':0.2}, ax=ax[2])
ax[2].set(ylim=(-3, 15))


plt.savefig('regplot_humidity.jpg', scale=4)
```

```python
fig, ax = plt.subplots(1,3,figsize=(19,6))

sns.regplot(x=df['prec-1'], y=df['prec'], color="b", marker=".", scatter_kws={'alpha':0.1}, ax=ax[0])
ax[0].set(ylim=(0, 25))

sns.regplot(x=df['prec-2'], y=df['prec'], color="b", marker=".", scatter_kws={'alpha':0.1}, ax=ax[1])
ax[1].set(ylim=(0, 25))

sns.regplot(x=df['prec-3'], y=df['prec'], color="b", marker=".", scatter_kws={'alpha':0.2}, ax=ax[2])
ax[2].set(ylim=(0, 25))


plt.savefig('regplot_prec.jpg', scale=4)
```

```python
# check correlation to discard columns if necessary
corr = df.corr()
cmap = sns.light_palette("#0c2a70", as_cmap=True)
mask = np.triu(corr)
plt.figure(figsize=(19,8))
sns.heatmap(corr, cmap='vlag', annot=True, linewidths=0.5, center=0.05, cbar=False, xticklabels=True,
           mask=mask)


plt.savefig('correlation.jpg', scale=4)
```
