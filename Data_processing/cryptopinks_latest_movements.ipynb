```python
import pandas as pd

from datetime import datetime
from datetime import date
import string
```

```python
df1=pd.read_csv('newcryptopunks.csv')
df2=pd.read_csv('newcryptopunks1.csv')
df3=pd.read_csv('newcryptopunks2.csv')
df4=pd.read_csv('newcryptopunks3.csv')
```

```python
df=pd.concat([df1,df2,df3,df4], axis=0).reset_index()
df=df.drop(columns=['Unnamed: 0','index'])
```

```python
df.head(15)
```

```python
df['fecha_venta'] = df['fecha_venta'].str.replace('none','')
df['fecha_apuesta'] = df['fecha_apuesta'].str.replace('none','')
df['fecha_oferta'] = df['fecha_oferta'].str.replace('none','')


df['fecha_venta'] = pd.to_datetime(df['fecha_venta'])
df['fecha_venta'] = pd.to_datetime(df['fecha_venta'], format="%d/%m/%Y")

df['fecha_apuesta'] = pd.to_datetime(df['fecha_apuesta'])
df['fecha_apuesta'] = pd.to_datetime(df['fecha_apuesta'], format="%d/%m/%Y")

df['fecha_oferta'] = pd.to_datetime(df['fecha_oferta'])
df['fecha_oferta'] = pd.to_datetime(df['fecha_oferta'], format="%d/%m/%Y")
```

```python
Split=df["venta"].str.split('(', expand=True)
Split.columns = ['venta_eth', 'venta_usd']
df = pd.concat([df, Split], axis=1)
```

```python
Split=df["apuesta"].str.split('(', expand=True)
Split.columns = ['apuesta_eth', 'apuesta_usd']
df = pd.concat([df, Split], axis=1)
```

```python
Split=df["oferta"].str.split('(', expand=True)
Split.columns = ['oferta_eth', 'oferta_usd']
df = pd.concat([df, Split], axis=1)
```

```python
df=df.drop(columns=['venta','apuesta','oferta'])
df.head(2)
```

```python
df['venta_eth']=df['venta_eth'].str.replace('Ξ','')
df['apuesta_eth']=df['apuesta_eth'].str.replace('Ξ','')
df['oferta_eth']=df['oferta_eth'].str.replace('Ξ','')
```

```python
df['venta_usd']=df['venta_usd'].str.replace(')','').str.replace(',','').str.replace('$','')
df['apuesta_usd']=df['apuesta_usd'].str.replace(')','').str.replace(',','').str.replace('$','')
df['oferta_usd']=df['oferta_usd'].str.replace(')','').str.replace(',','').str.replace('$','')
```
