```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('akumenius.txt',header=None,sep =" ",warn_bad_lines=True, error_bad_lines=False)
df
```

##### Eleminem algunes columnes

```python
columns_list = [0, 2, 3,5,7,8,9,11]
df = df.drop(columns=df.columns[columns_list])
df.tail(70)
```

```python

df.drop(df[df[10] == 'Apache (internal dummy connection)'].index, inplace = True)
df.drop(df[df[10] == 'Wget/1.12 (linux-gnu)'].index, inplace = True)
df[10] = df[10].str.replace(' http', ' +http')
```

```python
df.head()
```

```python
df[10].unique()
```

##### LLibreria ua_parser per extreure informaciÃ³ del user agent

```python
from ua_parser import user_agent_parser
ua = df[10].apply(lambda ua: user_agent_parser.Parse(ua))

df['device_brand'] = ua.apply(lambda x: x['device']['brand'])
df['device_model'] = ua.apply(lambda x: x['device']['model'])
df['os_family'] = ua.apply(lambda x: x['os']['family'])
df['user_agent_family'] = ua.apply(lambda x: x['user_agent']['family'])
```

```python
df.head()
```

```python
df[4] = df[4].str.replace('[', '')
df[6] = df[6].str.replace('GET /', '')
```

```python
df['user_agent_family'].unique()
```

##### canvi de nom de les columnes

```python
df.columns =['IP', 'date', 'URL', 'log','device_brand' , 'device_model' ,'os_family','user_agent_family']
df
```

##### conversÃ³ de la columna de la data a format datetime64

```python
df.date = df.date.str[:-9] + ' ' + df.date.str[-8:]
df['date']=df['date'].astype('datetime64')
```
